```python
X_file_name = 'day{}-data.txt'
YS_file_name = 'day{}-activities.txt'
YC_file_name = 'day{}-routines.txt'
DAYS = 7
input_dataset_path = "/Users/kavin/Silo/Research/WASH/complex_activities/Datasets/dataset_huynh_ubicomp08/"
output_dataset_path = "/Users/kavin/Silo/Research/WASH/Datasets/aroma_dataset_ubicomp08_filter_only_unlabeled/"
```

```python
def snip_dataset(inputs):
    for i in inputs:
        file_contents = []
        with open(input_dataset_path + i, 'r') as file:
            for c, l in enumerate(file):
                file_contents.append(l.strip())
            if(len(file_contents) % 750 != 0):
                with open(output_dataset_path + i, 'w') as f:
                    for line in file_contents[:(len(file_contents)//750)*750]:
                        f.write("%s\n" % line)
```

```python
X_INPUT_PATHS = [X_file_name.format(i + 1) for i in range(DAYS)]
YS_INPUT_PATHS = [YS_file_name.format(i + 1) for i in range(DAYS)]
YC_INPUT_PATHS = [YC_file_name.format(i + 1) for i in range(DAYS)]
```

```python
snip_dataset(X_INPUT_PATHS)
snip_dataset(YS_INPUT_PATHS)
snip_dataset(YC_INPUT_PATHS)
```

```python
#Filter only unlabeled rows
for day in range(DAYS):
    data_X = []
    data_YS = []
    data_YC = []
    bool_simple = []
    bool_complex = []
    with open(input_dataset_path + X_file_name.format(day + 1), 'r') as data_file:
        for c, l in enumerate(data_file):
            data_X.append(l.strip())
    with open(input_dataset_path + YS_file_name.format(day + 1), 'r') as simple_file:
        for c, l in enumerate(simple_file):
            data_YS.append(l.strip())
            if int(l.strip()) == 0:
                bool_simple.append(False)
            else:
                bool_simple.append(True)
    with open(input_dataset_path + YC_file_name.format(day + 1), 'r') as complex_file:
        for c, l in enumerate(complex_file):
            data_YC.append(l.strip())    
            if int(l.strip()) == 0:
                bool_complex.append(False)
            else:
                bool_complex.append(True)
    bool_X = [bs and bc for bs, bc in zip(bool_simple, bool_complex)]
    #print(bool_X)
    mod_data_X, mod_data_YS, mod_data_YC = [], [], []
    for index, line in enumerate(data_X):
        if(bool_X[index]):
            mod_data_X.append(line)
    for index, line in enumerate(data_YS):
        if(bool_X[index]):
            mod_data_YS.append(line)        
    for index, line in enumerate(data_YC):
        if(bool_X[index]):
            mod_data_YC.append(line)        
                    
    if(len(mod_data_X) % 750 != 0):
        with open(output_dataset_path + X_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_X[:(len(mod_data_X)//750)*750]):
                f.write("%s\n" % line)
        with open(output_dataset_path + YS_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_YS[:(len(mod_data_YS)//750)*750]):
                f.write("%s\n" % line)
        with open(output_dataset_path + YC_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_YC[:(len(mod_data_YC)//750)*750]):
                f.write("%s\n" % line)
    else:
        with open(output_dataset_path + X_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_X):
                f.write("%s\n" % line)
        with open(output_dataset_path + YS_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_YS):
                f.write("%s\n" % line)
        with open(output_dataset_path + YC_file_name.format(day + 1), 'w') as f:
            for index, line in enumerate(mod_data_YC):
                f.write("%s\n" % line)
```
